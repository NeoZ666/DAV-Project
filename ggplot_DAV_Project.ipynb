```r
# Scatter plot with additional options
library(ggplot2)
ggplot(data=mtcars, aes(x=wt, y=mpg)) +
  geom_point(pch=17, color="blue", size=2) +
  geom_smooth(method="lm", color="red", linetype=2) +
  labs(title="Automobile Data", x="Weight", y="Miles Per Gallon")
```

```r
# Scatter plot with faceting and grouping
data(mtcars)
mtcars$am <- factor(mtcars$am, levels=c(0,1),
                    labels=c("Automatic", "Manual"))
mtcars$vs <- factor(mtcars$vs, levels=c(0,1),
                    labels=c("V-Engine", "Straight Engine"))
mtcars$cyl <- factor(mtcars$cyl)

library(ggplot2)
ggplot(data=mtcars, aes(x=hp, y=mpg,
                        shape=cyl, color=cyl)) +
  geom_point(size=3) +
  facet_grid(am~vs) +
  labs(title="Automobile Data by Engine Type",
       x="Horsepower", y="Miles Per Gallon")
```

```r
library(ggplot2)
data(singer, package="lattice")

# Violin plot and boxplot
ggplot(singer, aes(x=voice.part, y=height)) +
  geom_violin(fill="lightblue") +
  geom_boxplot(fill="lightgreen", width=.2) +
  labs(title="Distribution of Height by Voice Part",
       x="Voice Part", y="Height") +
  theme_minimal()

# Density plot
ggplot(data=singer, aes(x=height, fill=voice.part)) +
  geom_density(alpha=.3) +
  labs(title="Density Plot of Height by Voice Part",
       x="Height", fill="Voice Part") +
  theme_minimal()

# Scatter plot
ggplot(singer, aes(x=voice.part, y=height)) +
  geom_point() +
  labs(title="Scatter Plot of Height by Voice Part",
       x="Voice Part", y="Height") +
  theme_minimal()

# Stacked bar plot
ggplot(singer, aes(x=voice.part, fill=voice.part)) +
  geom_bar() +
  labs(title="Stacked Bar Plot of Voice Part",
       x="Voice Part", fill="Voice Part") +
  theme_minimal()

# Dodged bar plot
ggplot(singer, aes(x=voice.part, fill=voice.part)) +
  geom_bar(position="dodge") +
  labs(title="Dodged Bar Plot of Voice Part",
       x="Voice Part", fill="Voice Part") +
  theme_minimal()

# Filled bar plot
ggplot(singer, aes(x=voice.part, fill=voice.part)) +
  geom_bar(position="fill") +
  labs(title="Filled Bar Plot of Voice Part",
       x="Voice Part", fill="Voice Part") +
  theme_minimal()
```

```r
# Placing options
library(ggplot2)

# Scatter plot with bars filled by sex
ggplot(mtcars, aes(x=as.factor(cyl), fill=factor(gear))) +
  geom_bar() +
  labs(title="Bar Plot of Cylinder Count by Gear Type", x="Cylinder Count", y="Count") +
  theme_minimal()

# Scatter plot with bars filled by red color
ggplot(mtcars, aes(x=as.factor(cyl))) +
  geom_bar(fill="red") +
  labs(title="Bar Plot of Cylinder Count", x="Cylinder Count", y="Count") +
  theme_minimal()

# Scatter plot with bars filled by red color (using aesthetic mapping)
ggplot(mtcars, aes(x=as.factor(cyl), fill="red")) +
  geom_bar() +
  labs(title="Bar Plot of Cylinder Count", x="Cylinder Count", y="Count") +
  theme_minimal()

data(singer, package="lattice")
library(ggplot2)
ggplot(data=singer, aes(x=height, fill=voice.part)) +
  geom_density() +
  facet_grid(voice.part~.)

# Faceting
data(singer, package="lattice")
library(ggplot2)
ggplot(data=singer, aes(x=height)) +
  geom_histogram() +
  facet_wrap(~voice.part, nrow=4)
```

```r
library(ggplot2)
ggplot(Salaries, aes(x=yrs.since.phd, y=salary, color=rank,
                     shape=rank)) + geom_point() + facet_grid(.~sex)

# Adding smoothed lines
data(Salaries, package="carData")
library(ggplot2)
ggplot(data=Salaries, aes(x=yrs.since.phd, y=salary)) +
  geom_smooth() + geom_point()

ggplot(data=Salaries, aes(x=yrs.since.phd, y=salary,
                          linetype=sex, shape=sex, color=sex)) +
  geom_smooth(method=lm, formula=y~poly(x,2),
              se=FALSE, size=1) +
  geom_point(size=2)

# Modifying axes
data(Salaries, package="carData")
library(ggplot2)
ggplot(data=Salaries, aes(x=rank, y=salary, fill=sex)) +
  geom_boxplot() +
  scale_x_discrete(breaks=c("AsstProf", "AssocProf", "Prof"),
                   labels=c("Assistant\nProfessor",
                            "Associate\nProfessor",
                            "Full\nProfessor")) +
  scale_y_continuous(breaks=c(50000, 100000, 150000, 200000),
                     labels=c("$50K", "$100K", "$150K", "$200K")) +
  labs(title="Faculty Salary by Rank and Sex", x="", y="")

# Legends
data(Salaries, package="carData")
library(ggplot2)
ggplot(data=Salaries, aes(x=rank, y=salary, fill=sex)) +
  geom_boxplot() +
  scale_x_discrete(breaks=c("AsstProf", "AssocProf", "Prof"),
                   labels=c("Assistant\nProfessor",
                            "Associate\nProfessor",
                            "Full\nProfessor")) +
  scale_y_continuous(breaks=c(50000, 100000, 150000, 200000),
                     labels=c("$50K", "$100K", "$150K", "$200K")) +
  labs(title="Faculty Salary by Rank and Gender",
       x="", y="", fill="Gender") +
  theme(legend.position=c(.1,.8))

# Scales
ggplot(mtcars, aes(x=wt, y=mpg, size=disp)) +
  geom_point(shape=21, color="black", fill="cornsilk") +
  labs(x="Weight", y="Miles Per Gallon",
       title="Bubble Chart", size="Engine\nDisplacement")

data(Salaries, package="carData")
ggplot(data=Salaries, aes(x=yrs.since.phd, y=salary, color=rank)) +
  scale_color_manual(values=c("orange", "olivedrab", "navy")) +
  geom_point(size=2)

ggplot(data=Salaries, aes(x=yrs.since.phd, y=salary, color=rank)) +
  scale_color_brewer(palette="Set1") + geom_point(size=2)

library(RColorBrewer)
display.brewer.all()
```

```r
library(ggplot2)

# Load sample dataset
data(mtcars)

# Plot heatmap
ggplot(mtcars, aes(x = factor(carb), y = factor(gear), fill = mpg)) +
  geom_tile() +
  scale_fill_gradient(low = "blue", high = "red") +
  labs(title = "Heatmap of mpg by carb and gear",
       x = "Carburetor", y = "Gear")
```

```r
library(igraph)
library(ggraph)

# Create sample network data
nodes <- data.frame(name = letters[1:5])
edges <- data.frame(from = c('a', 'b', 'c', 'd'), to = c('b', 'c', 'd', 'e'))

# Create igraph object
graph <- graph_from_data_frame(edges, directed = FALSE, vertices = nodes)

# Plot network graph
ggraph(graph, layout = "fr") +
  geom_edge_link() +
  geom_node_point() +
  geom_node_text(aes(label = name), vjust = 1)
```
